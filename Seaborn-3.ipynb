### Jointplot
jointplot create a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

```python
geyser = sns.load_dataset('geyser')
geyser.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser, hue='kind')
```

## Plot Kinds
There are currently six kinds of joint plots allowed in seaborn:
1. scatter(default)
2. kde
3. reg
4. hist
5. hex
6. resid

```python
sns.jointplot(x='waiting', y='duration', data=geyser, hue='kind', kind='hist')
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,  kind="reg", color='red' )
```

## FacetGrid

```python
sns.get_dataset_names()
```

```python
# load dataset
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['island'].unique()
```

```python
sns.set_style('white')
```

```python
sns.FacetGrid(penguins, col='island')
```

## .map()

```python
g = sns.FacetGrid(penguins, col='island')
g.map(sns.histplot, 'flipper_length_mm')
```

## .map_dataframe

```python
g = sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.histplot, x='flipper_length_mm', color='blue')
```

```python
g = sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.scatterplot, x='flipper_length_mm',y='bill_depth_mm', color='blue')
```

## .set_axis_labels()

```python
g = sns.FacetGrid(penguins, col='island', hue='species')
g.map_dataframe(sns.scatterplot, x='flipper_length_mm',y='bill_depth_mm')
g.set_axis_labels('flipper_length (mm)', 'bill_depth (mm)')
```

## PairGrid

```python
sns.set_style('dark')
g = sns.PairGrid(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map(sns.scatterplot)
```

## PairGrid.map_diag
## map_offdiag

```python
g = sns.PairGrid(penguins,hue='sex')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
```
